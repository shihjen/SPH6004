### Removing Records that contain Missing Values in any of the Attributes

We have identified a significant number of missing values within the dataset. Addressing missing values is crucial, and various approaches exist for handling them effectively. The most direct method involves removing records that contain missing values in any of the attributes. However, upon closer examination, we discovered that every record in the dataset possesses missing values in at least one attribute.

Executing a straightforward pd.DataFrame.dropna() operation would result in the removal of all records, rendering the dataset essentially empty. Therefore, a more sophisticated approach is imperative for effectively managing and imputing the missing values in a manner that preserves the integrity and utility of the dataset.

```python
import missingno as msno

# visualize the pattern of missingness in the training dataset
msno.matrix(train)
```

```python
# try with the straight forward approach to deal with missing value by removing all missing value in the dataset
# let's removing records that contain missing values in any of the attributes
train_copy = train.copy()
train_copy.dropna(inplace=True)

# check the dimension of the dataset after removing records contain missing values in any of the attributes
num_row = train_copy.shape[0]
print('By removing the records contain missing values in any of the attributes, the dataset left with {} records.'.format(num_row))
```

### Feature Selection

Given that a straightforward removal of records with missing values in any dataset attribute resulted in a complete absence of records, it becomes evident that a more systematic and analytical approach is essential for the feature selection process. Our strategy involves the identification of features or attributes exhibiting less than 30% missingness. Leveraging a tree-based approach, we aim to discern and prioritize the importance of these features, thus enabling a more refined and informed decision-making process for the subsequent analysis.

```python
# create a dataframe for contains information of missingness for each attributes in the dataset
missingValue_df = pd.DataFrame(train.isna().sum(), columns=['Num_Missing'])

# insert a new column with the percentage of missingness for each attribute in the dataset
missingValue_df['Percent_Missing'] = np.round((missingValue_df['Num_Missing'] / data.shape[0]) * 100, 2)
```

```python
# visualize the percentage of missingness for each attribute in the dataset
missingPercentage = missingValue_df['Percent_Missing'].sort_values(ascending=False)

# plot a horizontal bar chart
plt.figure(figsize=(12,30))
plt.barh(y=missingPercentage.index, width=missingPercentage.values, color='slateblue')
# add a line for 30% threshold for the missingness %
plt.axvline(x=30, color='red', linestyle='--', label='30% Threshold')
plt.xlabel('Percentage of Missingness (%)')

for index, values in enumerate(missingPercentage):
    plt.text(values+0.5, index, str(values), va='center')

plt.title('Percentage of Missingness of Each Attribute in the Dataset', fontsize=20)
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```

```python
# extract the variables in the dataset with percentage of missingness greater than 30%
lessMissingFeats = missingValue_df[missingValue_df['Percent_Missing'] >= 30].index.to_list()

# check the total number of variables that has less than 20% of missing value
print('There are {} variables in the training dataset have more than 30% of missing data. \n\nThese attributes are: {}'.format(len(lessMissingFeats), lessMissingFeats))
```

```python
# remove the variables containing more than 30% of missing data in training and test dataset
train.drop(columns=lessMissingFeats, axis=1, inplace=True)
test.drop(columns=lessMissingFeats, axis=1, inplace=True)

# check the dimension of the training and test dataset
print('Dimension of training dataset: ', train.shape)
print('Dimension of test dataset: ', test.shape)
```

### Helper Functions

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.feature_selection import SelectKBest, mutual_info_classif, SequentialFeatureSelector, RFE
from sklearn_genetic import GAFeatureSelectionCV

# helper function to calculate the information gain between each independent variable with target
# select top k features/independent variable with highest information gain with the target
def selectKBest_mi_feats(X,y,k=10):
    selector = SelectKBest(mutual_info_classif,k=10)
    selector.fit_transform(X,y)
    selected_feats = selector.get_feature_names_out()
    print('Features selected based on the information gain between independent variable with target:',', '.join(selected_feats))
    print()
    return selected_feats

# helper function to perform stepwise feature selection (forward or backward)
def sequentialFeatureSelection(X,y,C=0.1,num_feats=10,direction='forward'):
    estimator = LogisticRegression(penalty='l1',C=C,solver='saga',max_iter=500000,random_state=122)
    selector = SequentialFeatureSelector(estimator=estimator, n_features_to_select=num_feats, direction=direction)
    selector.fit(X,y)
    selected_feats = selector.get_feature_names_out()
    print('Features selected via stepwise forward selection approach:',', '.join(selected_feats))
    print()
    return selected_feats

# helper function to perform recursive feature elimination
def recursiveFeatureSelection(X,y,C=0.1,num_feats=10):
    estimator = LogisticRegression(penalty='l1',C=C,solver='saga',max_iter=500000,random_state=122)
    selector = RFE(estimator=estimator, n_features_to_select=num_feats)
    selector.fit(X,y)
    selected_feats = selector.get_feature_names_out()
    print('Features selected via recursive feature elimination approach:',', '.join(selected_feats))
    return selected_feats

# helper function to perform feature selection via tree-based model
def treeFeatSelector(estimator,X,y,topFeats=10,verbose=False):
    estimator.fit(X,y)
    featName = estimator.feature_names_in_
    score = estimator.feature_importances_
    impFeat = pd.Series(score, index=featName).sort_values(ascending=True)
    
    if verbose:
        plt.figure(figsize=(15,20))
        plt.barh(y=impFeat.index, width=impFeat.values, color='lightcoral')
        for index, values in enumerate (impFeat):
            plt.text(values+0.0005, index, str(round(values,4)), va='center')
        plt.title('Feature Importances from Tree-based Model', fontsize=20)
        plt.grid(alpha=0.4)
        plt.show()
    
    selected_feats = list(impFeat[-topFeats:].index)
    print('The top {} features selected via tree-based model are:'.format(topFeats),', '.join(selected_feats))
    return selected_feats

# helper function to perform feature selection via genetic algorithm
def gaFeatureSelection(estimator,X,y,population_size=10,generations=10):
    evolved_estimator = GAFeatureSelectionCV(
        estimator=estimator,
        cv=3,
        scoring='roc_auc',
        population_size=population_size,
        generations=generations,
        n_jobs=-1,
        verbose=True,
        keep_top_k=2,
        elitism=True,
    )

    evolved_estimator.fit(X,y)
    support = evolved_estimator.support_
    selected_feats = list(X.columns[support])
    print('\n{} features are selected via genetic algorithm.'.format(len(selected_feats)))
    print('These features are:',', '.join(selected_feats),'\n')
    return selected_feats
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

stratifiedCV = StratifiedKFold(n_splits=5)

# helper function to perform hyperparameters tuning via grid search and return the tuned model
def hyperparameterTuning(model,param_grid,selected_feats, Xtrain, ytrain):
    # create the GridSearchCV object
    grid_search = GridSearchCV(model, param_grid, cv=stratifiedCV, scoring='roc_auc', n_jobs=-1)

    # fit the grid search to the data
    grid_search.fit(Xtrain[selected_feats], ytrain)

    # print the best parameters and the corresponding accuracy
    print('Best Hyperparameters from Grid Search : ', grid_search.best_params_)
    print('Best ROC_AUC Score: ', grid_search.best_score_)
    print()

    # get the best model
    best_model = grid_search.best_estimator_
    
    # return the hyperparameters tuned model
    return best_model
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

# helper function to check model performance
# function to display classification report, ROC curve, and confusion matrix
def modelPerformance(model, Xtrain, Xtest, ytrain, ytest):
    ypred_train = model.predict(Xtrain)
    ypred_test = model.predict(Xtest)

    labels = ['No AKI','AKI']

    # classification report
    report_train = classification_report(ytrain, ypred_train, target_names=labels)
    report_test = classification_report(ytest, ypred_test, target_names=labels)

    print('Classification report for training data:')
    print(report_train)
    print('Classification report for test data:')
    print(report_test)

    # AUC-ROC curve
    fpr_train, tpr_train, _ = roc_curve(ytrain, ypred_train)
    roc_auc_train = auc(fpr_train, tpr_train)

    fpr_test, tpr_test, _ = roc_curve(ytest, ypred_test)
    roc_auc_test = auc(fpr_test, tpr_test)

    # plot ROC curve
    plt.figure(figsize=(10, 6))
    plt.plot(fpr_train, tpr_train, color='indigo', lw=2, label=f'Train ROC curve (AUC = {roc_auc_train:.2f})')
    plt.plot(fpr_test, tpr_test, color='lightcoral', lw=2, label=f'Test ROC curve (AUC = {roc_auc_test:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=20)
    plt.legend(loc='lower right')
    plt.show()
    
    # diagnosis - confusion matrix
    cm_train = confusion_matrix(ytrain, ypred_train, normalize='true')
    cm_test = confusion_matrix(ytest, ypred_test, normalize='true')

    figure, axes = plt.subplots(1,2, figsize=(13,5))
    sns.heatmap(cm_train, annot=True, cmap='BuPu', xticklabels=labels, yticklabels=labels, cbar=False, ax=axes[0])
    axes[0].set_title('Training', fontsize=15)
    sns.heatmap(cm_test, annot=True, cmap='BuPu', xticklabels=labels, yticklabels=labels, cbar=False, ax=axes[1])
    axes[1].set_title('Test', fontsize=15)
    figure.suptitle('Confusion Matrix', fontsize=20)
    plt.tight_layout(pad=2)
    plt.show()
```

```python
from scipy.stats import mannwhitneyu

# helper function to compute Cohen's d value
def cohen_d(group1, group2):
    mean_diff = np.mean(group1) - np.mean(group2)
    pooled_std = np.sqrt((np.std(group1, ddof=1) ** 2 + np.std(group2, ddof=1) ** 2) / 2)
    
    return mean_diff / pooled_std

# helper function to compute p-value and effect size of the features among 2 groups
def performStat(group1,group2,selected_feats,alternative='two-sided',effectSize=0.1):
    featDiff = []
    largeEffect = []
    test_stat_col = []
    pvalue_col = []
    effect_size_col = []
    for feat in selected_feats:
        res = mannwhitneyu(group1[feat], group2[feat], alternative=alternative)
        test_stat = res[0]
        test_stat_col.append(test_stat)
        pvalue = round(res[1],5)
        pvalue_col.append(pvalue)
        if pvalue > 0.05:
            featDiff.append(feat)
        # compute effect size (Cohen's d)
        effect_size = round(cohen_d(group1[feat],group2[feat]), 4)
        effect_size_col.append(effect_size)
        if abs(effect_size) > effectSize:
            largeEffect.append(feat)
            
    analysisResult = pd.DataFrame([test_stat_col,pvalue_col,effect_size_col]).T
    analysisResult.columns = ['Test Statistic', 'Test p-value','Effect Size']
    analysisResult.index = selected_feats
    display(analysisResult)
    print(f'{len(featDiff)} features have p-value greater than 0.05, which is not significant different between 2 groups.')
    print(f'{len(largeEffect)} features have effect size greater than {effectSize}.')
    return largeEffect
```

```python
def treeFeatSelector(estimator,X,y,topFeats=10,verbose=False):
    estimator.fit(X,y)
    featName = estimator.feature_names_in_
    score = estimator.feature_importances_
    impFeat = pd.Series(score, index=featName).sort_values(ascending=True)
    
    if verbose:
        plt.figure(figsize=(15,20))
        plt.barh(y=impFeat.index, width=impFeat.values, color='lightcoral')
        for index, values in enumerate (impFeat):
            plt.text(values+0.0005, index, str(round(values,4)), va='center')
        plt.title('Feature Importances from Tree-based Model', fontsize=20)
        plt.grid(alpha=0.4)
        plt.show()
    
    selected_feats = list(impFeat[-topFeats:].index)
    print('The top {} features selected via tree-based model are:',', '.join(selected_feats))
    return selected_feats
```

```python
# let's extract all the vital sign measurement (regardless of the min, max or mean value)
featName = data.columns.values

labFeatList = set()
demoList = set()
condition = ['_min','_max','_mean']

for feat in featName:
    if any(cond in feat for cond in condition):
        feat = feat.replace('_min','').replace('_max','').replace('_mean','')
        labFeatList.add(feat)
    else:
        demoList.add(feat)
        
print('Number of lab measurements: ', len(labFeatList))
print('Number of demographic and Glasgow Coma Scale measurements: ', len(demoList))
```

```python
# 'gcs_min' does not has a 'gcs_max' pair in the dataset
# remove from labFeatList
labFeatList.discard('gcs')

# insert into the demoList, with other Glascow Coma Scale measurements
demoList.add('gcs_min')
```

```python
# some of the lab measurements contain multiple measurements (.1, .2)
# get a list of these lab measurements
dotList = []
for feat in labFeatList:
    if '.' in feat:
        dotList.append(feat)        
print(dotList)
```

```python
# define a function to compute the mean for lab measurement
# the original min and max measurements will be removed
# return a copy of dataset with only mean value of the lab measurements, demographic data, and GCS data
def compute_mean(data, featList):
    data_copy = data.copy()
    for feat in featList:
        if feat in dotList:
            splitFeat = feat.split('.')
            featMean = splitFeat[0] + '_mean' + '.' + splitFeat[-1]
            featMax = splitFeat[0] + '_max' + '.' + splitFeat[-1]
            featMin = splitFeat[0] + '_min' + '.' + splitFeat[-1]
        else:
            featMean = feat + '_mean'
            featMax = feat + '_max'
            featMin = feat + '_min'
        data_copy[featMean] = (data_copy[featMax] + data_copy[featMin]) / 2
        data_copy.drop(columns=[featMax, featMin], axis=1, inplace=True)
    return data_copy
```

```python
# create a new dataset with only mean value of the lab measurements, demographic data, and GCS data
mean_data = compute_mean(data, list(labFeatList))
mean_data
```
